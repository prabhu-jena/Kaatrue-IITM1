```python
x=df.iloc[0: ,0:15]
```

```python
x.drop(x.iloc[:,1:5],axis=1,inplace=True)
```

```python
x.head()
```

```python
y=df.iloc[0:,15:]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
##  LINEAR REGRESSION
```

```python
from sklearn import linear_model
lin_reg = linear_model.LinearRegression()
lin_reg.fit(x_train,y_train)
print(lin_reg.score(x_train,y_train))
print(lin_reg.score(x_test,y_test))
lin_reg_pred=lin_reg.predict(x_test)
```

```python
lin_reg_pred
```

```python
from sklearn.metrics import r2_score
lin_reg_score=r2_score(y_test,lin_reg_pred)*100
print("Accuracy score for LR :",lin_reg_score)
```

```python
#####  RIDGE REGRESSION
```

```python
from sklearn.linear_model import Ridge
rr=Ridge()
rr
rr.fit(x_train,y_train)
print(rr.score(x_train,y_train))
print(rr.score(x_test,y_test))
rr_pred=rr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
rr_score=r2_score(y_test,rr_pred)*100
print("Accuracy score for RidgeR :",rr_score)
```

```python
##### LASSO REGRESSION
```

```python
from sklearn.linear_model import Lasso
lr=Lasso()
lr
lr.fit(x_train,y_train)
print(lr.score(x_train,y_train))
print(lr.score(x_test,y_test))
lr_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score

lr_score=r2_score(y_test,lr_pred)*100
print("Accuracy score for LR :",lr_score)
```

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x_train,y_train)
print(svr.score(x_train,y_train))
print(svr.score(x_test,y_test))
svr_pred=svr.predict(x_test)
from sklearn.metrics import r2_score
svr_score=r2_score(y_test,svr_pred)*100
print("Accuracy score for LR :",svr_score)
```

```python
lm = linear_model.LinearRegression()
lm.fit(x_train,y_train)
params = np.append(lm.intercept_,lm.coef_)
predictions = lm.predict(x_test)

newX = np.append(np.ones((len(x_train),1)), x_train, axis=1)
MSE = (sum((predictions)**2))/(len(newX)-len(newX[0]))

var_b = MSE*(np.linalg.inv(np.dot(newX.T,newX)).diagonal())
sd_b = np.sqrt(var_b)
ts_b = params/ sd_b

p_values =[2*(1-stats.t.cdf(np.abs(i),(len(newX)-len(newX[0])))) for i in ts_b]

sd_b = np.round(sd_b,3)
ts_b = np.round(ts_b,3)
p_values = np.round(p_values,3)
params = np.round(params,4)

myDF3 = pd.DataFrame()
myDF3["Coefficients"],myDF3["Standard Errors"],myDF3["t values"],myDF3["Probabilities"] = [params,sd_b,ts_b,p_values]
print(myDF3)
```
